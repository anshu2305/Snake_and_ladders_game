```python
from PIL import Image
import random
end=100

def check_ladder(p):
    if(p==2):
        print("Ladder")
        return 23
    elif (p==6):
        print("Ladder")
        return 45
    elif (p==20):
        print("Ladder")
        return 59
    elif (p==52):
        print("Ladder")
        return 72
    elif (p==57):
        print("Ladder")
        return 96
    elif(p==71):
        print("Ladder")
        return 92
    else:
        return p

def check_snake(p):
    if (p==43):
        print("Snake")
        return 17
    elif (p==50):
        print("Snake")
        return 5
    elif (p==56):
        print("Snake")
        return 8
    elif (p==73):
        print("Snake")
        return 15
    elif (p==84):
        print("Snake")
        return 63
    elif (p==87):
        print("Snake")
        return 49
    elif (p==98):
        print("Snake")
        return 40
    else:
        return p
    
def show_board():
    img=Image.open('desktop/snake.jpg')
    img.show()

def play():
    p1=input("Player1,Enter youe name: ")
    p2=input("Player2,Enter your name: ")
    pp1=0
    pp2=0
    turn=0
    while(1):
        if(turn%2==0):
            print("")
            print("Its the turn of ",p1)
            c=input("Press 1 to continue the game and 0 to exit : ")
            print("")
            if c=='0':
                    print("")
                    print(p1," score is:",pp1)
                    print("")
                    print(p2," score is:",pp2)
                    print("")
                    print("Thankyou for playing!!")
                    break
            dice=random.randint(1,6)
            print(p1,"was at",pp1)
            print("Dice showed : ",dice)
            pp1=pp1+dice
            pp1=check_ladder(pp1)
            pp1=check_snake(pp1)
            print(p1,"is at",pp1)
            if pp1>end:
                pp1==end
                print("")
                print(p1,"Won!!")
        else:
            print("")
            print("Its the turn of ",p2)
            c=input("Press 1 to continue the game and 0 to exit : ")
            print("")
            if c=='0':
                print("")
                print(p1," score is:",pp1)
                print("")
                print(p2," score is:",pp2)
                print("")
                print("Thankyou for playing!!")
                break
            dice=random.randint(1,6)
            print(p2,"was at",pp2)
            print("Dice showed : ",dice)
            pp2=pp2+dice
            pp2=check_ladder(pp2)
            pp2=check_snake(pp2)
            print(p2,"is at",pp2)
            if pp2>end:
                pp2==end
                print("")
                print(p2,"won!!")
        turn=turn+1    
show_board()
play()  
```
